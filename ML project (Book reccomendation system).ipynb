## POPULARITY BASED RECCOMENDATION

```python
ratings
```

```python
ratings_with_title = ratings.merge(books,on='ISBN')
```

```python
ratings_with_title
```

```python
print(ratings_with_title.shape)
```

```python
ratings_with_title.groupby('Book-Title').count()['Book-Rating'].reset_index() ##here book-rating is number of times the book has been rated. 
```

```python
num_ratings_df = ratings_with_title.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_ratings_df.rename(columns={'Book-Rating' : 'Num-Ratings'},inplace=True)
```

```python
num_ratings_df
```

```python
ratings_with_title.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
```

```python
avg_rating_df = ratings_with_title.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating' : 'Average-Rating'}, inplace=True)
```

```python
avg_rating_df
```

```python
ratings_df = num_ratings_df.merge(avg_rating_df,on='Book-Title')
```

```python
ratings_df
```

```python
ratings_df = ratings_df.sort_values('Average-Rating', ascending=False)
```

```python
ratings_df
```

```python
## Popular books top 50.
popular_df = ratings_df[ratings_df['Num-Ratings']>=250].head(50)
```

```python
popular_df
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Num-Ratings','Average-Rating','Image-URL-L']]
```

```python
popular_df
```

## COLLABORATIVE BASED RECCOMENDER SYSTEM

```python
t = ratings_with_title.groupby('User-ID').count()['Book-Rating']>=200
```

```python
t ## shows us whether a user had given rating to at least 200 books or not.
```

```python
t[t] ## gives us users who had give rating to atleast 200 books
```

```python
u200 = t[t].index ## fetchs the indexs of the users who had give rating to atleast 200 books.
```

```python
u200
```

```python
filtered_ratings = ratings_with_title[ratings_with_title['User-ID'].isin(u200)]  ## gives us df of users who had given rating to atleast 200 books.
```

```python
filtered_ratings
```

```python
z = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50  ##shows us whether a book has atleast 50 or not.
```

```python
z
```

```python
z[z] ## gives the books which has atleast 50 ratings.
```

```python
b50 = z[z].index ## fetches the indexs of above.
```

```python
b50
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(b50)] ## Gives us df of the book which have atleast 50 ratings.
```

```python
final_ratings
```

```python
ub = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating') ##Gives us final table of books vs User-ID over rating.
```

```python
ub
```

```python
ub.fillna(0,inplace=True) ## Turns nan into 0.
```

```python
ub
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(ub)
```

```python
similarity_score.shape ## Give the simlarity of the book with the rest of the books.
```

```python
    def r(book_name):
        #fetching index from book name.
        index = np.where(ub.index==book_name)[0][0]
        #fetching similar items
        similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:6]

        for i in similar_items:
            print(ub.index[i[0]])
```
